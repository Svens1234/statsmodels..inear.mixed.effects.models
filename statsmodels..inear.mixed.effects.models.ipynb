```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
data = sm.datasets.get_rdataset("dietox", "geepack").data
```

```python
data
```

```python
md = smf.mixedlm("Weight ~ Time", data, groups=data["Pig"])
```

```python
mdf = md.fit()
```

```python
mdf.summary()
```
